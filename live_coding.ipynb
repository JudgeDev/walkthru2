# A simple system for annotating and note taking

I have been following the fast.ai project for some time now. I went through an early version of the course and used some of the content to enter an internal work competition with some colleagues to classify patent documents. We did moderately well, but with room for improvement.

What struck me, as many have noted, was the quantity and quality of the material. I did not think of making my own notes on the lectures at that time, but I did attempt to draw a large diagram of the buildup of the concepts in part two of the course.

I have recently come back to the course and stumbled on the Live Coding videos, initially called Walkthrus. In my opinion, this work rasises the bar to an extrodinary level not only in terms of quantity of the content, but in teaching workflow, especially efficiency.

It is stated in the book [https://www.amazon.de/-/en/Steve-McConnell/dp/0735619670](Code Complete) that the productivity of individual programmers varies by a factor of 10 to 20. I also read that startup programmers are often an order of magnitude more productive than even good programmers. Jeremy Howard is clearly that type of programmer. What is incredible, however, is that he is willing and manages to explain many aspects of his know-how in these videos.

The practical result is that there is so much information that it is difficult to do it justice. One approach is to keep re-watching the videos and following along until the techniques become ingrained. However, at may age, that is proving to be a race against time.

Another approach is obviously to take notes, but how?

There is a lot of information about note taking in connection with the "second brain" concept. But there does not seem to be very much information on the syntax of the notes themselves. Is seems to be assumed that the notes are plain language sentences, possibly with some shortening.

At (https://forums.fast.ai/t/live-coding-5/96733/35#35)[one point] in the discussion of Live Coding 5, Daniel asks, "I wonder could you show us how you take note with obsidian on walkthrough videos, especially with your style. By style, I mean what are your useful tricks and tips in how to effectively use internal links, tags, and vim binding keys, etc to note down ideas/questions, to organise and to search the entire notebooks." This got me thinking.

In the course of my career I have been annotating documents, summarising documents, taking notes, writing, and giving presentations for a long time. I have only just realised how badly. Badly, because, although I follow a broadly similar approach it is not very rigorous.

So I thought I'd jot down some of the conventions that I use. This is partly for my own benefit, but also it might be useful as a basis for a more general system that can be improved by other people.

On the Live Coding pages, Daniel produces detailled notes with associated timestamps on the videos. 

They seem to be in the form:
```xml
<time> "-" <Heading>
"How to x?" <command>
```

My inital format was more like code:
```xml
<command>  "#" <comment>
```
or user manual style:
```xml
"Jupyter" ">" <menu> ">" <menu-item>
```

In fact I seem to be using the convention:

| Symbol | Meaning |
|--------|---------|
| >      | consecutive steps on an interface |
|\#      | explantation after code|
| -      | explanation except after code |
| ->     | result of step/code or conclusion in argument (see below) |
| (...)  | optional
| \|     | or |
